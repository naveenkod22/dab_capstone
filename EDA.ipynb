```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Data Understanding:

```python
replay = pd.read_csv('./replay.csv')
sybil = pd.read_csv('./replay_sybil.csv')
```

```python
replay.plot.scatter(x = 'pos_x', y = 'pos_y')
plt.show()
```

```python
sns.jointplot(data = replay.drop_duplicates('sender'), x = 'pos_x', y = 'pos_y', kind='hex', marginal_kws=dict(bins=30, fill=True))
plt.xlabel('Position X')
plt.ylabel('Position Y')
plt.show()
```

```python
sns.jointplot(data = replay.drop_duplicates('sender'), x = 'spd_x', y = 'spd_y', kind='hex', marginal_kws=dict(bins=30, fill=True))
plt.xlabel('Speed X')
plt.ylabel('Speed Y')
plt.grid()
plt.show()
```

```python
sns.jointplot(data = replay.drop_duplicates('sender'), x = 'acl_x', y = 'acl_y', kind='hex', marginal_kws=dict(bins=30, fill=True))
plt.xlabel('Accleration X')
plt.ylabel('Accleration Y')
plt.grid()
plt.show()
```

```python
sns.jointplot(data = replay.drop_duplicates('sender'), x = 'hed_x', y = 'hed_y', kind='hex', marginal_kws=dict(bins=30, fill=True))
plt.xlabel('Head Position X')
plt.ylabel('Head Position Y')
plt.grid()
plt.show()
```
