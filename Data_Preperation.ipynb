```python
import json
import glob 
import pandas as pd
from pathlib import Path
```

```python
file_names = glob.glob(".\VeReMi_25200_28800_2019-11-27_16_36_54\*.json")

data = pd.DataFrame()
for i in range(1, len(file_names)):
    file_name =  file_names[i]
    df = pd.read_json(file_name, lines=True)
    df['file_name'] = file_name
    data = data.append(df)
   
data = data[data.columns.drop(list(data.filter(regex='noise')))]

data[['pos_x','pos_y', 'pos_z']] = pd.DataFrame(data['pos'].to_list(), columns=['pos_x','pos_y', 'pos_z'])
data[['spd_x','spd_y','spd_z']] = pd.DataFrame(data['spd'].to_list(), columns=['spd_x','spd_y','spd_z'])
data[['acl_x','acl_y', 'acl_z']] = pd.DataFrame(data['acl'].to_list(), columns=['acl_x','acl_y', 'acl_z'])
data[['hed_x','hed_y', 'hed_z']] = pd.DataFrame(data['hed'].to_list(), columns=['hed_x','hed_y', 'hed_z'])

data['Type'] =  data.file_name.str[-15:-13]

data.drop(columns=['pos', 'spd', 'acl', 'hed', 'file_name'], inplace=True)
```
