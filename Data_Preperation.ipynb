```python
import pandas as pd
```

# Preparing replay attack data:

```python
# Reading two data files and storing them into one dataframe called replay.
df1 = pd.read_csv('./data/DataReplay_1416/DataReplay_1416_50400_54000_2019-11-28_10_46_53.csv')
df2 = pd.read_csv('./data/DataReplay_1416/VeReMi_54000_57600_2019-11-28_10_46_53.csv')
replay = df1.append(df2,  ignore_index=True)
replay.head().T
```

```python
cols = replay.columns[replay.nunique()==1]
replay.drop(columns=cols, inplace=True)
replay = replay.loc[:,~replay.columns.str.contains('noise')]
replay.drop(columns= ['file_name','sendTime','senderPseudo'], inplace= True)
replay.rename(columns= {'attack?':'attack'}, inplace = True)
replay.columns
```

# Preparing replay sybil attack data:

```python
df1 = pd.read_csv('./data/DataReplaySybil_1416/DataReplaySybil_1416_VeReMi_50400_54000_2019-11-28_10_43_31.csv')
df2 = pd.read_csv('./data/DataReplaySybil_1416/DataReplaySybil_1416_VeReMi_54000_57600_2019-11-28_10_43_31.csv')
sybil = df1.append(df2,  ignore_index=True)
sybil.head()
```

```python
cols = sybil.columns[sybil.nunique()==1]
sybil.drop(columns=cols, inplace=True)
sybil = sybil.loc[:,~sybil.columns.str.contains('noise')]
sybil.drop(columns= ['file_name','sendTime'], inplace= True)
sybil.rename(columns= {'attack?':'attack'}, inplace = True)
sybil.columns
```

# Writing dataframes to csv files.

```python
replay.to_csv('replay.csv', index=False)
sybil.to_csv('replay_sybil.csv', index = False)
```
